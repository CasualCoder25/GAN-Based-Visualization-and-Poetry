```python
# Import dependencies
from diffusers import DiffusionPipeline
import torch
```

```python
vision_gen_model = "openskyml/midjourney-mini"
```

```python
pipe = DiffusionPipeline.from_pretrained(
    vision_gen_model,
    cache_dir="./model/",
).to("cuda")
```

```python
# Setup prompt
prompt = "painting of a group of people walking down a street with umbrellas"
```

```python
image = pipe(prompt=prompt)
```
